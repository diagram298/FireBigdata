```python
import pandas as pd
report1 = pd.read_csv('original/소방청_소소심 익히기 소방안전교육 현황_20211231.csv')
report2 = pd.read_excel('original/시도별_산업별_총괄_20230125133820.xlsx')
report3 = pd.read_excel('original/시도별노후건축물수_2021.xlsx')
# report4 = pd.read_csv('original/예산~~~~.csv')
```

```python
# 소방청_소소심 익히기 소방안전교육 현황_20211231
SAFETY_TRAINING = pd.DataFrame()
SAFETY_TRAINING['REGION_NAME'] = report1.본부별
SAFETY_TRAINING['YEAR'] = 2021
SAFETY_TRAINING['TRAINED_PERSON'] = report1['유아 인원'] + report1['초등학생(저) 인원'] + report1['초등학생(고) 인원'] + report1['중고등학생 인원'] + report1['성인 인원'] + report1['노인 인원'] + report1['장애인 인원'] + report1['외국인 인원']
SAFETY_TRAINING['CATEGORY_NAME'] = '화재'

SAFETY_TRAINING.to_csv('data/SAFETY_TRAINING.csv', index=False)
SAFETY_TRAINING
```

```python
# 시도별 노후건축물 수
report3 = report3.dropna()
OLD_BUILDING = pd.DataFrame()
OLD_BUILDING['REGION_NAME'] = report3.구분[1:]
OLD_BUILDING['YEAR'] = 2021
OLD_BUILDING['BUILDING_NUM'] = report3.합계[1:]
OLD_BUILDING['CATEGORY_NAME'] = '화재'

OLD_BUILDING.to_csv('data/OLD_BUILDING.csv', index=False)
OLD_BUILDING
```

```python
# 시도별 창고/운송업 수

report2 = report2.fillna(method='ffill')
report2 = report2[report2.산업별 == '창고 및 운송관련 서비스업']

WAREHOUSE_TRANSIT = pd.DataFrame()
WAREHOUSE_TRANSIT['REGION_NAME'] = report2.시도별[1:]
WAREHOUSE_TRANSIT['YEAR'] = 2020
WAREHOUSE_TRANSIT['SERVICE_NUM'] = report2['2020'][1:]
WAREHOUSE_TRANSIT['CATEGORY_NAME'] = '화재'

WAREHOUSE_TRANSIT.to_csv('data/WAREHOUSE_TRANSIT.csv', index=False)
WAREHOUSE_TRANSIT
```
