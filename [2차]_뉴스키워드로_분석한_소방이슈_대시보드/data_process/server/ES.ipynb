```python
import pandas as pd
from pororo import Pororo
```

```python
zsl = Pororo(task="zero-topic", lang="ko")
```

```python
df = pd.read_csv('ES_main.csv',encoding='utf-8-sig')
```

```python
df = df[['YNewsML.NewsContent.Title','YNewsML.NewsContent.Body','send_timestamp']]
```

```python
df.columns
```

```python
import numpy as np
```

```python
df['P-category'] = np.nan
```

```python
for i in range(len(df['YNewsML.NewsContent.Title'])):
    df['P-category'][i] = [df['YNewsML.NewsContent.Title'][i], zsl(df['YNewsML.NewsContent.Title'][i],['정치','경제','사회','문화','스포츠','국제','지방','IT/과학','북한','생활','경제','금융','스포츠'])]
```

```python
df['P-category'][2][1]
```

```python
df
```

```python
ca =[]
for i in range(len(df)):
        ca.append(pd.DataFrame(df['P-category'][i][1],index=[0]).T.reset_index().sort_values(by=0).rename(columns={'index':'cate'}).reset_index().drop('index',axis=1)['cate'][0])
```

```python
pd.DataFrame(ca).rename(columns={0:'cate'}).to_csv('ES_cate.csv',encoding='utf-8-sig',index=False)
```
