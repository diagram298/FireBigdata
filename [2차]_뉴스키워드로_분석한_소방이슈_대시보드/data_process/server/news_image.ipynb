```python
df = pd.read_csv('../Final_fire_news.csv')
```

```python
link_list = df['YNewsML.Metadata.Href'].tolist()
len(link_list)
```

```python
# Use BS

from bs4 import BeautifulSoup as bs
from urllib import request

img_link_bs = []
for i in link_list:
    try:
        html = request.urlopen(i)
        soup = bs(html, 'html.parser')
        a = soup.select_one('#articleWrap > div.content01.scroll-article-zone01 > div > div > article > div.comp-box.photo-group > figure > div > span > img').get('src')
        img_link_bs.append(a[2:])
    except:
        try:
            html = request.urlopen(i)
            soup = bs(html, 'html.parser')
            a = soup.select_one('#viewWrap > div.article-img > img').get('src')
            img_link_bs.append(a[2:])
        except:
            img_link_bs.append('')
img_link_bs
```

```python
df['img_link'] = img_link_bs
df
```

```python
df.to_csv('../NEWS_NLP/Final_fire_news_2.csv',index=False)
```
