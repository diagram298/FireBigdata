```python
pick = df[(df['YNewsML.NewsContent.Title'].str.contains('종합5보'))]
pick
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
b = pd.DataFrame()

for i in pick['YNewsML.NewsContent.Title']:
    a = []
    c = ''
    for result, score in kiwi.analyze(i, top_n=1):
        for j in range(len(result)):
            if result[j][1] == 'NNG':
                c += result[j][0] + ', '
    a.append(c)
    b = pd.concat([b,pd.DataFrame([a])],axis=0)
b
```

```python
pd.DataFrame(b[0][0].split(', ')[:-1]).drop_duplicates().reset_index().drop('index',axis=1)
```

```python
b_li = b[0].tolist()
spell_list = []
for i in b_li:
    spell = []
    text = i.split(', ')
    for j in text:
        if (len(j) > 1) & (j != '종합') :
            spell.append(j)
    spell_list.append(spell)
spell_list
```

```python
from itertools import combinations
aa = list(combinations(spell_list[0],2))
aa
```

```python
# 검색
df_frame = pd.DataFrame()
for i in aa: 
    cond1 = df['YNewsML.NewsContent.Body'].str.contains(i[0])
    cond2 = df['YNewsML.NewsContent.Body'].str.contains(i[1])
#     cond3 = df['YNewsML.NewsContent.Body'].str.contains(i[2])
    ddff = df[cond1&cond2]
    df_frame = pd.concat([df_frame, ddff])
    
table1 = df_frame.drop_duplicates()
table1
```
