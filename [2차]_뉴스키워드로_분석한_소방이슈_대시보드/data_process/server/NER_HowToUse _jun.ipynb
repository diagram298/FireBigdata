```python
from NER import  NerFactory
import pandas as pd
```

## 사고

```python
df = pd.read_excel('/home/intern2/fire_dev/NEWS_NLP/bigkinds/NewsResult_20220203-20230203 (3).xlsx')
```

```python
df.columns
```

```python
df1 = df['제목']
```

```python
ner = NerFactory()
```

```python
sentence = df['제목'][:500].tolist()
```

```python
a = []
b = []
for title in df1 :
    result = ner.get_entities(title)
    a.append(title)
    b.append(result)
```

```python
big = pd.DataFrame([a,b]).T.rename(columns={0:"title",1:"NER"})
```

```python
EV_word_big = []
for i in range(len(big)):
    for j in range(len(big['NER'][i])):
        if big['NER'][i][j]['code'] == 'EV':
            EV_word_big.append([big['NER'][i][j]['word'],big['title'][i]])
```

```python
EV_word_big
```

```python
big_result = pd.DataFrame(pd.DataFrame(EV_word_big)).rename(columns={0:'word',1:'title'})
big_result
```

```python
big_result.to_csv('big.csv',encoding='utf-8-sig',index=False)
```
