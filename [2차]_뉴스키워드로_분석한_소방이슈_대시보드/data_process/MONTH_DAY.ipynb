```python
import pandas as pd
MONTH1 = pd.read_csv('original/시도별인구현황.csv')
DAY1 = pd.read_csv('original/시도별화재인명피해현황.csv')
DAY2_1 = pd.read_csv('original/소방안전신문고_2023_JAN_day.csv')
DAY2_2 = pd.read_csv('original/소방안전신문고_2023_FEB_1to6_day.csv')
report = pd.read_csv('original/report.csv')
```

```python
# 지역별 화재건수, 희생자 수 (API)
FIRE = pd.DataFrame()
FIRE['REGION_NAME'] = DAY1.SIDO_NM
FIRE['DATE'] = DAY1.OCRN_YMD
FIRE['FIRE_NUM'] = DAY1.OCRN_MNB
FIRE['DEAD_NUM'] = DAY1.VCTM_PERCNT
FIRE['CATEGORY_NAME'] = '화재'

FIRE.to_csv('data/FIRE.csv', index=False)
FIRE
```

```python
# 지역별 인구수 (API)
MONTH1 = MONTH1.astype({'수록시점':'str'})
POPULATION = pd.DataFrame()
POPULATION['REGION_NAME'] = MONTH1['분류값명1']
POPULATION['YEAR'] = MONTH1['수록시점'].str[:4]
POPULATION['MONTH'] = MONTH1['수록시점'].str[4:]
POPULATION['POPULATION'] = MONTH1.수치값
POPULATION['CATEGORY_NAME'] = '화재'

POPULATION.to_csv('data/POPULATION.csv', index=False)
POPULATION
```

```python
# 지역별 소방안전신문고
SAFETY_REPORT = pd.DataFrame(columns=['REGION_NAME','DATE','REPORT_NUM','CATEGORY_NAME'])

for i in range(31):
    df = pd.DataFrame()
    df['REGION_NAME'] = DAY2_1.region
    df['DATE'] = DAY2_1.columns[i+1]
    df['REPORT_NUM'] = DAY2_1.iloc[:,i+1]
    df['CATEGORY_NAME'] = '화재'
    SAFETY_REPORT = pd.concat([SAFETY_REPORT,df])

for i in range(6):
    df = pd.DataFrame()
    df['REGION_NAME'] = DAY2_2.region
    df['DATE'] = DAY2_2.columns[i+1]
    df['REPORT_NUM'] = DAY2_2.iloc[:,i+1]
    df['CATEGORY_NAME'] = '화재'
    SAFETY_REPORT = pd.concat([SAFETY_REPORT,df])

SAFETY_REPORT.to_csv('data/SAFETY_REPORT.csv', index=False)
SAFETY_REPORT
```
